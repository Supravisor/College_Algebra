# Function and a table of (x,y) values

```python
x = 5
y = 4*x + 3 # The function 

print(x,",",y)

print("x \t y") 

for x in range(11):
    y = 4*x + 3  # The function
    print(x, "\t", y)
```

# Defining a Python function

```python
def f(x):
   y = 4*x + 3
   return y

print(5, ",", f(5)) 
```
