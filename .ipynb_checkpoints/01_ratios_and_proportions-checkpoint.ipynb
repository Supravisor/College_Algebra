# Set up a proportion:<br>
$\frac{n1}{d1}=\frac{n2}{d2}$

```python
# Put a zero in for the unknown value
n1 = 1
d1 = 2
n2 = 4
d2 = 0
```

```python
if n2==0:
    answer = d2 * n1 / d1
    print("n2 = ", answer)
```

```python
if d2==0:
    answer = n2 * d1 / n1
    print("d2 = ", answer)
```

# Solving in other ways

```python
%pip install sympy
from sympy import *

var('x y')

# First equation set equal to zero, ready to solve
first = 2*x - y


# Sympy syntax for equation equal to zero, ready to factor
eq1 = Eq(first,0)

# Sympy solve for x
sol = solve(eq1,x)

# Show factored results
print("x = ", sol[0])
```

# Explaining how each function works

## converts string input (including fractions) to float

```python
def string_frac(in_string):
    if "/" in in_string:
        nd = in_string.split("/")
        n = float(nd[0])
        d = float(nd[1])
        ans = n/d
        return ans
    else:
        ans = float(in_string)
        return ans
string_frac("1/2")
```

## simplest one-step addition

```python
def one_step_add():
    import random
    # Display problem
    a = random.randint(-4,10)
    b = random.randint(2,24)
    print("x + ", a, " = ", b)
    ans = float(input("x = "))
    answer = b-a
    # Test input
    if ans==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
one_step_add()
```

## one-step additon with negative numbers

```python
def one_step_subtract():
    import random
    a = random.randint(-19,-1)
    b = random.randint(2,24)
    print(a, " + x = ", b)
    ans = float(input("x = "))
    # test
    answer = b-a
    if ans==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
one_step_subtract()
```

## one-step multiply

```python
def one_step_mult():
    # Uses string_frac(<input string>)
    import random
    a = random.randint(1,11)
    b = random.randint(2,24)
    print(a, "x = ", b)
    ans_in = (input("x = "))
    answer = round(b/a,2)
    # test
    if round(string_frac(ans_in),2)==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
one_step_mult()
```

## one-step divide

```python
def one_step_div():
    import random
    a = random.randint(1,11)
    b = random.randint(2,24)
    print("x/", a, " = ", b)
    ans = float(input("x = "))
    answer = b*a
    # test
    if ans==answer:
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
one_step_div()
```

## two-step problems

```python
def two_step():
    import random
    # Uses string_frac()
    a = random.randint(1,11)
    b = random.randint(-7,12)
    c = random.randint(2,36)
    print(a, "x + ", b, " = ", c)
    print("Round answer to two decimal places")
    ans_in = input("x = ")
    answer = (c-b)/a
    # test
    if round(string_frac(ans_in),2)==round(answer,2):
        print("Correct! \n")
    else:
        print("Try again")
        print("The correct answer is ", answer, "\n")
two_step()
```
