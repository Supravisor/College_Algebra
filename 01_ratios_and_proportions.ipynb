Set up a proportion:<br>
$\frac{n1}{d1}=\frac{n2}{d2}$

```python
# Put a zero in for the unknown value
n1 = 1
d1 = 2
n2 = 4
d2 = 0
```

```python
if n2==0:
    answer = d2 * n1 / d1
    print("n2 = ", answer)
```

```python
if d2==0:
    answer = n2 * d1 / n1
    print("d2 = ", answer)
```

Doing more with the solution

```python
%pip install sympy
import sympy
from sympy import symbols
from sympy.solvers import solve

x = symbols('x')

# Put the equation here
eq = 2*x - 4

solution = solve(eq,x)
print("x = ", solution[0])
```

Multiple answers

```python
%pip install sympy
from sympy import symbols
from sympy.solvers import solve

x = symbols('x')

eq = input('Enter equation: 0 = ')

solution = solve(eq,x)
for s in solution:
    print("x = ", s)
```
