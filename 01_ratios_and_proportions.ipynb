Set up a proportion:<br>
$\frac{n1}{d1}=\frac{n2}{d2}$

```python
# Put a zero in for the unknown value
n1 = 1
d1 = 2
n2 = 4
d2 = 0
```

```python
if n2==0:
    answer = d2 * n1 / d1
    print("n2 = ", answer)
```

```python
if d2==0:
    answer = n2 * d1 / n1
    print("d2 = ", answer)
```

```python
%pip install sympy
from sympy import symbols
from sympy.solvers import solve

x = symbols('x')

eq = input('Enter equation: 0 = ')

solution = solve(eq,x)
for s in solution:
    print("x = ", s)
```

Explaining how each function works

```python
# converts string input (including fractions) to float
def string_frac(in_string):
    if "/" in in_string:
        nd = in_string.split("/")
        n = float(nd[0])
        d = float(nd[1])
        ans = n/d
        return ans
    else:
        ans = float(in_string)
        return ans
string_frac("1/2")
```
