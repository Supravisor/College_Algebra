# Code for slope intercept

```python
x1 = 1
y1 = 7
x2 = 2
y2 = 10

slope = (y2 - y1) / (x2 - x1)

print("slope = ", slope)
```

```python
x1 = 632 #1
y1 = 84 #7
x2 = 1800 #2
y2 = 8 #10

# The slope is "m"
m = (y2 - y1) / (x2 - x1)

# The y intercept is "b"
b = y1 - m*x1

# The full equation
print("y = ", m, "x + ", b)
```
