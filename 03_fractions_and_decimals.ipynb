# Positive and negative exponents

```python
print(10**1)
print(10**2)
print(10**3)

print(10**0)

print(10**-1)
print(10**-2)
print(10**-3)
```

# User input comes in as a string. You can find the length of the string

```python
#User input comes in as a string. You can find the length of the string
text = input("Enter a number")
print(text)

print(len(text))
```

# Cast input as a float or int variable

```python
text = input("Enter a number")
num = float(text)
print(num + 4)
```
